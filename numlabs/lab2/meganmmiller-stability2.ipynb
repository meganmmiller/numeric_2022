```python
# Import
import matplotlib.pyplot as plt
from lab2_functions import euler,beuler,leapfrog
import numpy as np
```

```python
# Define
theFuncs={'euler':euler,'beuler':beuler,'leapfrog':leapfrog}
```

<h1> Forward Euler </h1>

```python
# Plot
if __name__=="__main__":
    tend=10.
    Ta=20.
    To=30.
    theLambda=-8.
    funChoice='euler'
    npts=np.array([39, 40,41])
    
      
    for n in npts:
        n_string = str(n)
        delta_t = str(round(tend/n,3))
    
        approxTime,approxTemp=theFuncs[funChoice](n,tend,To,Ta,theLambda)
        exactTime=np.empty([n,],float)
        exactTemp=np.empty_like(exactTime)
        for i in np.arange(0,n):
           exactTime[i] = tend*i/n
           exactTemp[i] = Ta + (To-Ta)*np.exp(theLambda*exactTime[i])
        plt.close('all')
        plt.figure(1)
        plt.clf()
        plt.plot(exactTime,exactTemp,'r+')
        plt.plot(approxTime,approxTemp)
        theAx=plt.gca()
        theAx.set_xlim([0,10])
        theAx.set_ylim([0,45])
        theAx.set_title("Stability for " + n_string +" points or \u0394t =" +delta_t )
        plt.show()
```

<h2>Comments on Forward Euler Graphs</h2>

The plots above show that the value Δ𝑡<−2/𝜆 does in fact predict the onset of stability for the forward euler scheme. 

The first plot above ("Stability for 39 points or Δ𝑡=0.256") shows a case when Δ𝑡>−2/𝜆; as can be seen in the plot the oscillations around the exact solution are increasing with time, which indicates an unstable condition.  

The middle plot ("Stability for 40 points or Δ𝑡=0.25") shows a condition where Δ𝑡=−2/𝜆; as can be seen in the plot the oscillations in the numerical solution remain constant around the exact solution with time.  

In the third plot ("Stability for 41 points or Δ𝑡=0.244") Δ𝑡<−2/𝜆, and the oscillations in the numerical solution converge on the exact solution with time; which indicates a stable condition.

<h1> Backward Euler </h1>

```python
# Plot
if __name__=="__main__":
    multiplier =4
    tend=10*multiplier
    Ta=20.
    To=30.
    theLambda=-8.
    funChoice='beuler'
    npts=np.array([multiplier*39, multiplier*40,multiplier*41])
    
      
    for n in npts:
        n_string = str(n)
        delta_t = str(round(tend/n,3))
    
        approxTime,approxTemp=theFuncs[funChoice](n,tend,To,Ta,theLambda)
        exactTime=np.empty([n,],float)
        exactTemp=np.empty_like(exactTime)
        for i in np.arange(0,n):
           exactTime[i] = tend*i/n
           exactTemp[i] = Ta + (To-Ta)*np.exp(theLambda*exactTime[i])
        plt.close('all')
        plt.figure(1)
        plt.clf()
        plt.plot(exactTime,exactTemp,'r+')
        plt.plot(approxTime,approxTemp)
        theAx=plt.gca()
        theAx.set_xlim([0,tend])
        theAx.set_ylim([19,21])
        theAx.set_title("Stability for " + n_string +" points or \u0394t =" +delta_t )
        plt.show()
```

<h2>Comments on Backward Euler Graphs</h2>
   
For the three values of Δ𝑡 plotted the backward euler numeric solution quickly nears the exact solution and does not show any signs of instability.  The solution was tested to a time of 100 seconds, with no observed instability.

This is consistent with the backward euler stability problem in Lab 2 which indicates that this scheme is stable for all cases when 𝜆 is real and negative.

<h1> Leap-Frog</h1>

```python
# Plot
if __name__=="__main__":
    tend=10.
    Ta=20.
    To=30.
    theLambda=-8.
    funChoice='leapfrog'
    npts=np.array([39, 40,41])
    
      
    for n in npts:
        n_string = str(n)
        delta_t = str(round(tend/n,3))
    
        approxTime,approxTemp=theFuncs[funChoice](n,tend,To,Ta,theLambda)
        exactTime=np.empty([n,],float)
        exactTemp=np.empty_like(exactTime)
        for i in np.arange(0,n):
           exactTime[i] = tend*i/n
           exactTemp[i] = Ta + (To-Ta)*np.exp(theLambda*exactTime[i])
        plt.close('all')
        plt.figure(1)
        plt.clf()
        plt.plot(exactTime,exactTemp,'r+')
        plt.plot(approxTime,approxTemp)
        theAx=plt.gca()
        theAx.set_xlim([0,tend])
        theAx.set_ylim([-10000,10100])
        theAx.set_title("Stability for " + n_string +" points or \u0394t =" +delta_t )
        plt.show()
```

<h2>Comments on Leap-Frog Graphs</h2>
   
For the three values of Δ𝑡 plotted, the leap-frog numeric solution had expanding oscillations with an amplitude greater than 20000  before time equal to 2 seconds. This indicates that the leap-frog scheme is not stable for these conditions.

The bottom graph with the lowest Δ𝑡 does show a slight reduction in the oscillations as the predicted value for the 5th time step is only ~-7500 whereas for the top two graphs it is less than -10,000. However, the solution is still unstable.
